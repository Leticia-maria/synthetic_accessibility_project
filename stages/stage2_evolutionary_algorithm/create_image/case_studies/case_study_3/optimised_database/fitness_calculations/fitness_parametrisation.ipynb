```python
from pathlib import Path
import stk
import json
from rdkit.Chem import 
%autoreload 2
```

```python
pop_paths = list(Path('/Users/stevenbennett/PhD/main_projects/synthetic_accessibility_project/stages/stage2_evolutionary_algorithm/create_image/case_studies/case_study_3/optimised_database/opt_pops').glob('pop*/pop_fitness.json'))
pops = []
for i in pop_paths:
    pops.append(stk.Population.load(str(i)))
```

```python
pops[8][0].sc_score
```

```python
counter = 0
collapsed = []
uncollapsed = []
for pop in pops:
    for mem in pop:
        if not mem.pore_diameter or mem.pore_diameter < 1:
            collapsed.append(mem)
            print('Cage collapsed.')
            counter += 1
        elif mem.pore_diameter > 5:
            uncollapsed.append(mem)
```

```python
print(counter)
```

```python
tot = 0
for i in pops:
    tot += len(i)
print(f'There are {tot} cages.')
print(f'{(counter/tot)*100} of these cages have collapsed.')
```
